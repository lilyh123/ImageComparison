```python
import cv2 # a model of opencv
from google.colab import files # library that allows the user to upload files
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
from PIL import Image

def orb_sim(img1, img2):
  orb = cv2.ORB_create()

  # detect keypoints and descriptors
  keypoint_a, description_a = orb.detectAndCompute(img1, None)
  keypoint_b, description_b = orb.detectAndCompute(img2, None)

  # define the bruteforce matcher object
  bruteforce = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)

  # perform matches
  matches = bruteforce.match(description_a, description_b)
  # look for similar regions with distance < 67 
  similar_regions = [i for i in matches if i.distance < 67]
  if len(matches) == 0:
    return 0
  return len(similar_regions) / len(matches)

# store the similarity values
totalSimilarity = 0

for x in range(1, 4):
  for y in range(1, 4):
    image1 = 'rey_complex_0' + str(x) + '_0' + str(y) + '.png'
    image2 = 'drawing1_0' + str(x) + '_0' + str(y) + '.png'

    rey_complex = cv2.imread(image1, 0)
    drawing = cv2.imread(image2, 0)

    # show the image
    title1 = image1 + ' Image'
    plt.title(title1)
    image = mpimg.imread(image1)
    plt.imshow(image)
    plt.show()
    title2 = image2 + ' Image'
    plt.title(image2)
    image = mpimg.imread(image2)
    plt.imshow(image)
    plt.show()

    # find the similarity value
    orb_similarity = orb_sim(rey_complex, drawing)
    totalSimilarity += orb_similarity
    print("Similarity using ORB is: ", orb_similarity)

print("Average similarity using ORB is: ", totalSimilarity/9)
```
